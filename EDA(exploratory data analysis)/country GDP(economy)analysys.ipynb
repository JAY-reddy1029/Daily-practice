```python
vis4 = sns.boxplot(data = df, x="IncomeGroup", y="BirthRate" )  #boxplot
plt.show(vis4)
```

```python
vis5 = sns.lmplot(data = df, x="InternetUsers", y="BirthRate" ) # lmplot
plt.show(vis5)
```
